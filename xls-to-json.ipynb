```python
import pandas as pd
import json

# Load the Excel file
df = pd.read_excel('HSK词汇下载202502201435.xls')

# Select only the columns 0, 1, 2, and -1
df_selected = df.iloc[:, [0, 1, 2, -1]]

# Rename the columns to match the format you need
df_selected.columns = ['char', 'pinyin', 'definition', 'hsk']

# Convert to a list of dictionaries (one dictionary per row)
result = df_selected.to_dict(orient='records')

# Convert the list of dictionaries to a JSON string
json_string = json.dumps(result, ensure_ascii=False, indent=4)

# Print the JSON string
print(json_string)
```
